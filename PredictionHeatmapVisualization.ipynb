## Dependencies

```python
import os
import cv2
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
```

## Generating Heatmap based on predictions

```python
img_filepath = 'Data/NewsOutlet/NewsOutlet-TE/images/'
pred_filepath = 'Predictions/NewsOutlet/RGB_VST/'
img_filenames = os.listdir(img_filepath)
pred_filenames = os.listdir(pred_filepath)
N = len(pred_filenames)

fig, axes = plt.subplots(math.ceil(N/5), 5,figsize=(25, 5 * int(math.ceil(N/5))))
ax = axes.ravel()

for image_i in range(N):
    img_filename = img_filenames[image_i]
    img_path = img_filepath + img_filename
    pred_path = pred_filepath + img_filename[:-3] + 'png'

    # receiving image as variables im and pred
    im = cv2.imread(img_path)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    pred = cv2.imread(pred_path)
    pred = cv2.cvtColor(pred, cv2.COLOR_BGR2RGB)


    pred_heatmap = cv2.applyColorMap(pred, cv2.COLORMAP_HOT)
    axes[math.floor(image_i/5),image_i%5].imshow(im, alpha=0.5)
    axes[math.floor(image_i/5),image_i%5].imshow(pred_heatmap, alpha=0.6)
    

# Show plots
for a in ax:
    a.set_axis_off()

# Save heatmaps as high quality pdf
fig.savefig('NewsOutlet.pdf', format='pdf', dpi=600)
plt.show()
```
